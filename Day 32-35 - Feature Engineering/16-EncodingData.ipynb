```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic') 
```

```python
df
```

```python
df[["sex", "class", "embark_town"]].isna().sum()
```

```python
df = df.dropna(subset=['embark_town'])
```

```python
df
```

```python
df[["sex", "class", "embark_town"]].isna().sum()
```

```python
# one hot encoding
```

```python
df['sex'].value_counts()
```

```python
df.embark_town.value_counts()
```

```python
df['class'].value_counts()
```

```python
df_onehot = pd.get_dummies(df, columns=['sex', 'embark_town'], drop_first=True)
```

```python
df_onehot
```

```python
# label encoder
```

```python
label_encoder = LabelEncoder()
```

```python
df_label = df.copy()
```

```python
df_label
```

```python
df_label['sex'] = label_encoder.fit_transform(df_label['sex'])
```

```python
df_label
```

```python
# ordinal encoding
```

```python
df_ordinal = df.copy()
```

```python
#class_order = ["Third", "Second", "First"]  # Define order
class_order = ["First" , "Second" , "Third"]  # olmuyor
```

```python
ordinal_encoder = OrdinalEncoder(categories=[class_order])
```

```python
df_ordinal['class'] = ordinal_encoder.fit_transform(df_ordinal[['class']])
```

```python
df_ordinal
```

```python
# Visualization to see the effect of encoding
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
df["sex"].value_counts().plot(kind="bar", ax=axes[0], title="Original Categorical")
df_onehot[["sex_male"]].sum().plot(kind="bar", ax=axes[1], title="One-Hot Encoded")
df_label["sex"].value_counts().plot(kind="bar", ax=axes[2], title="Label Encoded")
```
