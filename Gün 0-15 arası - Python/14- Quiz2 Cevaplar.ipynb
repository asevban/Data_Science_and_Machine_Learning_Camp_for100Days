```python
#7) Aşağıdaki listedeki sayılar bir dairenin yarı çapını vermektedir. 
```

```python
#Tüm dairelerin çevresini içeren başka yeni bir liste oluşturunuz. (İpucu: 2 * pi * r)  Pi 3.14 alınabilir.
```

```python
r_list = [3,2,5,8,4,6,9,12]
```

```python
c_list = []
```

```python
for r in r_list:
    cevre = 2 * 3.14 * r
    c_list.append(cevre)
c_list
```
