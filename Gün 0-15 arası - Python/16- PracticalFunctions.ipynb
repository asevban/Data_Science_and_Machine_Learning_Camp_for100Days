```python
def divideNumber(number):
    return number / 2
```

```python
divideNumber(20)
```

```python
myList = [3,5,7,10,20,30]
```

```python
my_result_list = []
```

```python
for num in myList:
    my_result_list.append(divideNumber(num))
print(my_result_list)
```

```python
def controlString(string):
    return "Atil" in string
```

```python
controlString("Atil Samancioglu")
```

```python
controlString("Samancioglu")
```

```python
myStringList = ["Atil", "Atil Samancioglu","Samancioglu","Atlas"]
```

```python
list(map(controlString,myStringList))
```

# Filter

```python
list(filter(controlString,myStringList))
```

# LEGB -> L->Local, E-> Enclosing, G->Global, B->Built-In

```python
#Global
myString = "Atıl"

def myFunction():
    #Enclosing
    myString = "Atıl 2"
    print(myString)
    
    def myFunction2():
        #Local
        myString = "Atıl 3"
        print(myString)
    
    myFunction2()
```

```python
myFunction()
```
