```python
my_list = [10,20,30,40,50,60,70]
```

```python
print("for loop started")
for num in my_list:
    print(num)
print("for loop ended")
```

```python
for number in my_list:
    print(number)
    if number == 40:
        print("40 bulundu")
        break
```

```python
for number in my_list:
    if number == 40:
        continue
    print(number)
```

```python
for num in my_list:
    pass
```
