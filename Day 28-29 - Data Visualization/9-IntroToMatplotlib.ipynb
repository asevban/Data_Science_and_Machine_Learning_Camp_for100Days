```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline  # notebooklarda bu çalıştımalısın önce
```

```python
age_list = [10,20,30,30,30,40,50,60,70,75]
```

```python
weight_list = [20,60,80,85,86,87,70,90,95,90]
```

```python
plt.plot(age_list,weight_list, 'b')   #y -> yellow, g,b,r ...
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
numpy_arr1 = np.linspace(0,10,20)
```

```python
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1 ** 3
```

```python
numpy_arr2
```

```python
plt.plot(numpy_arr1,numpy_arr2, 'bo-')
plt.xlabel('x')
plt.ylabel('x ** 3')
plt.title('x vs x**3')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arr1,numpy_arr2,"ro-")
plt.subplot(1,2,2) # 1 row, 2 columns, 2nd graph
plt.plot(numpy_arr2,numpy_arr1,"go-")
plt.show()
```
