```python
# z-score
```

```python
data = np.array([10, 12, 13, 15, 18, 25, 100, 105])
```

```python
# outlier
```

```python
mean = np.mean(data)
```

```python
mean
```

```python
data.sum() / len(data)
```

```python
std = np.std(data)
```

```python
std
```

```python
z_scores = (data - mean) / std
```

```python
z_scores
```

```python
np.abs(z_scores) > 1
```

```python
data[z_scores > 1]
```

```python
# Solve 2x + 3y = 8 and 5x + 7y = 19
```

```python
# Coefficients matrix
A = np.array([[2, 3], [5, 7]])
```

```python
# Constants matrix
b = np.array([8, 19])
```

```python
# Solve for x and y
solution = np.linalg.solve(A, b)
```
