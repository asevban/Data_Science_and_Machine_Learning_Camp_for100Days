# inheritance

```python
class Musician():
    
    def __init__(self,name):
        self.name = name
        print("musician class")
    
    def test1(self):
        print("test1")
        
    def test2(self):
        print("test2")
```

```python
atil = Musician("atil")
```

```python
atil.name
```

```python
atil.test1()
```

```python
atil.test2()
```

```python
class MusicianPlus(Musician):     #miras aldı

    def __init__(self,name):
        Musician.__init__(self,name)    #miras alınan sınıfla aynı
        print("musician plus")
    
    def test3(self):
        print("test3")

    def test1(self):
        print("test1 test1 test1")
```

```python
atlas = MusicianPlus("atlas")
```

```python
atlas.test1()
```

```python
atlas.test2()
```

```python
atlas.test3()
```

```python
type(atil)
```

```python
atil.test1()
```

```python
atlas.test1()
```

# Polymorphism

```python
class Banana():
    
    def __init__(self,name):
        self.name = name

    def info(self):
        return f"100 calories {self.name}"

class Apple():
    
    def __init__(self,name):
        self.name = name

    def info(self):
        return f"150 calories {self.name}"
```

```python
banana = Banana("banana")
```

```python
apple = Apple("apple")
```

```python
banana.info()
```

```python
apple.info()
```

```python
fruitList = [banana,apple]
```

```python
for fruit in fruitList:
    print(fruit.info())
```

# Encapsulation

```python
class Phone():

    def __init__(self, name, price):
        self.name = name
        self.__price = price  # __ önüne bunları koyarak değiştirilmez hale getiriyorum

    def info(self):
        print(f"{self.name} price is: {self.__price}")

    def changePrice(self, price):
        self.__price = price
```

```python
iphone = Phone("iphone" , 500)
```

```python
iphone.info()
```

```python
iphone.__price = 400
```

```python
iphone.info()
```

```python
iphone.changePrice(300)
```

```python
iphone.info()
```

# Abstraction

```python
from abc import ABC, abstractmethod
```

```python
class Car(ABC):
    @abstractmethod
    def maxSpeed(self):
        pass
```

```python
#myCar = Car()
```

```python
class Tesla(Car):
    
    def maxSpeed(self):
        print("200km")        
```

```python
tesla = Tesla()
```

```python
tesla.maxSpeed()
```

```python
class Fruit():
    
    def __init__(self,name,calories):
        self.name = name
        self.calories = calories

    def __str__(self):
        return f"{self.name}:{self.calories} calories"

    def __len__(self):
        return self.calories
```

```python
myFruit = Fruit("banana" , 150)
```

```python
print(myFruit)
```

```python
len(myFruit)
```
