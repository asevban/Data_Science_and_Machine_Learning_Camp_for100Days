```python
def salary_category(salary):
    if salary < 5000:
        return "Low"
    elif 5000 <= salary < 8000:
        return "Medium"
    else:
        return "High"
```

```python
df["Salary_Category"] = df["Salary"].apply(salary_category)
```

```python
df
```

```python
def new_perf_score(row):
    if row["Experience"] > 10:
        return row["Performance_Score"] + 1
    else:
        return row["Performance_Score"]
```

```python
df["new_perf_score"] = df.apply(new_perf_score , axis=1)
```

```python
df
```

```python
df["Formatted_Name"] = df["Name"].apply(lambda x: x.replace("_", " "))
```

```python
# Satış departmanında 5 yıldan fazla deneyimi olanlar

sales_senior = df[ (df['Department'] == 'Sales')  &  (df['Experience'] > 5) ]
```

```python
# Performans skoru 4 ve 5 olanlar (yüksek performans)

perf_skor = df[df['Experience'] >= 5]
```

```python
perf_skor
```

```python
def experience_level(years):
    if years < 3:
        return "Junior"
    elif years < 7:
        return "mid"
    else:
        return "Senior"
```

```python
df['Experience_Level'] = df['Experience'].apply(experience_level)
```
